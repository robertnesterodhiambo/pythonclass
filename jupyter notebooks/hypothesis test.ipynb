# 1.
- On average, a certain pesticide contains 2% toxic material. Batches of the pesticide that contain too little or too much of the material cannot be approved for retail sale. In a random sample of 80 batches of the pesticide, the mean amount of toxic material was 1.93% with a standard deviation of 0.2% (or 0.0193 and 0.002). Do the data indicate that the mean amount of toxic material in the pesticide differs from 2%? Use a significance level of 5% for the hypothesis test AND a 95% confidence interval to justify your claim.

1. **Hypothesis Test:**

   Test Statistic:
   $ z = \frac{{\bar{x} - \mu}}{{\frac{{\sigma}}{{\sqrt{n}}}}} $
   
   Given:
   - Sample mean (\( \bar{x} \)): 1.93%
   - Population mean (\( \mu \)): 2%
   - Population standard deviation (\( \sigma \)): 0.002
   - Sample size (\( n \)): 80

   Calculate:
   $ z = \frac{{1.93 - 2}}{{\frac{{0.002}}{{\sqrt{80}}}}} $
   $ z = \frac{{-0.07}}{{\frac{{0.002}}{{8.944}}}} $
   $ z = \frac{{-0.07}}{{0.000224}} $
   $ z = -312.5 $

   Conclusion:
   Since \( |z| > 1.96 \), we reject the null hypothesis.

2. **95% Confidence Interval:**

   Confidence Interval:
   $ \text{Confidence interval} = \bar{x} \pm z \times \frac{{\sigma}}{{\sqrt{n}}} $

   Given:
   - Sample mean (\( \bar{x} \)): 1.93%
   - Population standard deviation (\( \sigma \)): 0.002
   - Sample size (\( n \)): 80
   - Critical value at 95% confidence level (from Z-table): ±1.96

   Calculate:
   $ \text{Confidence interval} = 1.93 \pm 1.96 \times \frac{{0.002}}{{\sqrt{80}}} $
   $ \text{Confidence interval} = 1.93 \pm 1.96 \times 0.000224 $
   $ \text{Confidence interval} = 1.93 \pm 0.00043984 $

   Therefore, the 95% confidence interval is \( (1.92956\%, 1.93044\%) \).

   Therefore, both the hypothesis test and the confidence interval analysis provide evidence that the mean amount of toxic material in the pesticide differs from 2% at the 5% significance level.

# 2.
 - The Association of Oklahoma Independent Oil Producers would like to prove that the average production of Oklahoma oil wells is less than 24 barrels per day. The association randomly sampled seventeen actively pumping oil wells in the state. The sample produced a mean of 20.8 barrels per day and a standard deviation of 3.6 barrels. Use a 0.01 significance level to test the Associations claim. 


1. **Hypothesis Test:**

   Null Hypothesis (H0): $ \mu \geq 24 $
   
   Alternative Hypothesis (H1): $ \mu < 24 $
   
   Test Statistic: $ t = \frac{{\bar{x} - \mu}}{{\frac{{s}}{{\sqrt{n}}}}} $
   
   Given:
   - Sample mean ($ \bar{x} $): 20.8
   - Population mean ($ \mu $): 24
   - Sample standard deviation ($ s $): 3.6
   - Sample size ($ n $): 17
   
   Calculate:
   $ t = \frac{{20.8 - 24}}{{\frac{{3.6}}{{\sqrt{17}}}}} $
   $ t = \frac{{-3.2}}{{\frac{{3.6}}{{4.123}}}} $
   $ t = \frac{{-3.2}}{{0.871}} $
   $ t \approx -3.670 $

2. **Critical Value:**
   
   For a one-tailed test with a significance level of 0.01 and degrees of freedom ($ df $) equal to 16, the critical value is approximately -2.583 (from a t-distribution table).

3. **Conclusion:**
   
   Since the calculated test statistic ($ t $) of approximately -3.670 is less than the critical value of -2.583, we reject the null hypothesis.
   
   Therefore, there is sufficient evidence to support the Association's claim that the average production of Oklahoma oil wells is less than 24 barrels per day at the 0.01 significance level.

   Null Hypothesis (H0): $ \mu_{\text{Texas}} = \mu_{\text{Illinois}} $ (There is no significant difference in the acidity rates between Texas and Illinois)
   
   Alternative Hypothesis (H1): $ \mu_{\text{Texas}} \neq \mu_{\text{Illinois}} $ (There is a significant difference in the acidity rates between Texas and Illinois)

   Significance Level: $ \alpha = 0.05 $

2. **Calculate Test Statistic:**
   
   We will use the formula for the two-sample t-test for independent samples:
   $ t = \frac{{\bar{x}_1 - \bar{x}_2}}{{\sqrt{\frac{{s_p^2}}{{n_1}} + \frac{{s_p^2}}{{n_2}}}}}$
   where $ \bar{x}_1 $ and $ \bar{x}_2 $ are the sample means, $ n_1 $ and $ n_2 $ are the sample sizes, and $ s_p^2 $ is the pooled variance.

3. **Calculate Pooled Variance ($ s_p^2 $):**
   
   $ s_p^2 = \frac{{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}}{{n_1 + n_2 - 2}} $
   where $ s_1^2 $ and $ s_2^2 $ are the sample variances.

4. **Find Degrees of Freedom ($ df $):**
   
   Degrees of freedom for the two-sample t-test:
   $ df = n_1 + n_2 - 2 $


calculations:

1. **Calculate Sample Means ($ \bar{x}_1 $ and $ \bar{x}_2 $):**

   For Texas:
   $ \bar{x}_1 = \frac{{4.69 + 5.10 + 5.22 + 4.46 + 4.93 + 4.65 + 5.22 + 4.76 + 4.25 + 5.14}}{{10}} \\ $
   $ \bar{x}_1 = \frac{{49.42}}{{10}} \\  $

   $ \bar{x}_2 = \frac{{4.40 + 4.69 + 4.22 + 4.64 + 4.54 + 4.35 + 4.69 + 4.40 + 4.75 + 4.93 + 4.45 + 4.49}}{{12}} \\ $
   $ \bar{x}_2 = \frac{{53.1}}{{12}} \\ $
   $ \bar{x}_2 = 4.425 \\ $

2. **Calculate Sample Variances ($ s_1^2 $ and $ s_2^2 $):**

   For Texas:
   $ s_1^2 = \frac{{\sum (x - \bar{x}_1)^2}}{{n_1 - 1}} \\ $
   $ s_1^2 = \frac{{(4.69 - 4.942)^2 + (5.10 - 4.942)^2 + \ldots + (5.14 - 4.942)^2}}{{10 - 1}} \\ $
   $ s_1^2 = \frac{{0.0301 + 0.0264 + \ldots + 0.0209}}{{9}} \\ $
   $ s_1^2 \approx \frac{{0.2168}}{{9}} \\ $
   $ s_1^2 \approx 0.0241 \\ $

   For Illinois:
   $ s_2^2 = \frac{{\sum (x - \bar{x}_2)^2}}{{n_2 - 1}} \\  $
   $ s_2^2 = \frac{{(4.40 - 4.425)^2 + (4.69 - 4.425)^2 + \ldots + (4.49 - 4.425)^2}}{{12 - 1}} \\  $

3. **Calculate Pooled Variance ($ s_p^2 $):**
   
   $ s_p^2 = \frac{{(10 - 1) \times 0.0241 + (12 - 1) \times 0.0275}}{{10 + 12 - 2}} \\  $
   $ s_p^2 = \frac{{9 \times 0.0241 + 11 \times 0.0275}}{{20}} \\  $

   $ s_p^2 \approx \frac{{0.5194}}{{20}} \\  $
   $ s_p^2 \approx 0.02597 \\  $

4. **Find Degrees of Freedom ($ df $):**
   
   $ df = n_1 + n_2 - 2 \\ \\  $
   $ df = 10 + 12 - 2 = 20 \\ \\  $

5. **Find Critical Value:**
   
   For a two-tailed test with a significance level of 0.05 and degrees of freedom ($ df $) equal to 20, the critical value can be found using a t-distribution table or a statistical software. Let's assume the critical value to be approximately ±2.086 (based on a t-distribution table).

6. **Calculate Test Statistic ($ t $):**
   
   $ t = \frac{{\bar{x}_1 - \bar{x}_2}}{{\sqrt{\frac{{s_p^2}}{{n_1}} + \frac{{s_p^2}}{{n_2}}}}} \\  $
   $ t = \frac{{4.942 - 4.425}}{{\sqrt{\frac{{0.02597}}{{10}} + \frac{{0.02597}}{{12}}}}} \\  $
